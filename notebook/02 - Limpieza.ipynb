Siguiendo con la limpieza de datos, la variable "Age - group" debemos mejorarla. Ya que mezcla edades numéricas, rangos de edad, incluso persenta literales como 'Age not stated'.

También deberemos normalizar de alguna manera la variable "Area Name" para quedarme solo con el nombre del estado.

Empezamos con la variable "Age-group":

```python
print("Valores únicos de Age-group:")
print(df_Educ_clean['Age-group'].unique())

print("\nValores únicos de Area Name:")
print(df_Educ_clean['Area Name'].unique())
```

```python
## 1. Normalizamos Area Name:

df_Educ_clean['Area Name'] = (
    df_Educ_clean['Area Name']
    .str.replace('State - ', '', regex=False)  # quita el prefijo
    .str.replace(r'\s+\d+$', '', regex=True)   # quita el número del final
    .str.strip()                               # quita espacios sobrantes
)

print(df_Educ_clean['Area Name'].unique())
```

```python
## 2. Normalizamos Age-group generando rangos de edad:

def clasificar_edad(x):
    if x == 'Age not stated':
        return 'desconocido'
    elif x == '0-6':
        return 'infancia'
    elif x in ['7','8','9','10','11','12','13','14']:
        return 'ninez'
    elif x in ['15','16','17','18','19']:
        return 'adolescencia'
    elif x in ['20-24','25-29']:
        return 'juventud'
    elif x in ['30-34','35-39','40-44','45-49']:
        return 'adulto'
    elif x in ['50-54','55-59','60-64']:
        return 'adulto_mayor'
    else:  # 65-69, 70-74, 75-79, 80+
        return 'tercera_edad'

df_Educ_clean['age_group'] = df_Educ_clean['Age-group'].apply(clasificar_edad)

print(df_Educ_clean['age_group'].value_counts())
```

# Fusión de los dos df: df_Educ_clean y df_Rank_final2

Tal como dijimos anteriormente, para obtener una métrica representativa del ranking de cada estado para cruzarla con el dataframe de educación (df_Educ_clean), voy a calcular el promedio de los 5 años. 

Para ello, creo un nuevo dataframe "df_Rank_state" a partir de df_Rank_final2

```python
# Agrupar ranking por estado (promedio de los 5 años)
df_Rank_state = df_Rank_final2.groupby('State').agg(
    Score_mean=('Score', 'mean'),
    Score_std=('Score', 'std'),
    Rank_mean=('Rank', 'mean'),
    TLR_mean=('TLR', 'mean'),
    RPC_mean=('RPC', 'mean'),
    GO_mean=('GO', 'mean'),
    OI_mean=('OI', 'mean'),
    PERCEPTION_mean=('PERCEPTION', 'mean'),
    n_institutos=('Institute ID', 'count')  # me devuelve el total de univ. por Estado
).reset_index()

print(df_Rank_state.shape)  # Esto me devolverá 23 estados
df_Rank_state.head()
```

Ultimo paso: Como dijimos también anteriormente, necesito tener el df_Educ_clean a nivel estado, es decir, un estado por fila, con el fin de fusionarlo con el df_Rank_state.

Agregar df_Educ_clean a nivel estado (es eliminar Age_group implícitamente) para obtener así una fila por estado
y poderlo fusionar con df_Rank_state.

```python
df_Educ_state = (
    df_Educ_clean
    .groupby("Area Name")
    .sum(numeric_only=True)
    .reset_index()
)
```

```python
df_Educ_clean.head()
```

```python
df_Educ_state.head()
```

```python
# Creo columnas adicionales con la tasa educativa. Para ello, convierto a porcentajes las variables numéricas,

# Lista de columnas de educación
columns = [
"Illiterate - Persons",
"Illiterate - Males",
"Illiterate - Females",
"Literate - Persons",
"Literate - Males",
"Literate - Females",
"Literate w/o education level - Persons",
"Literate w/o education level - Males",
"Literate w/o education level - Females",
"Bellow primary - Persons",
"Bellow primary - Males",
"Bellow primary - Females",
"Primary - Persons",
"Primary - Males",
"Primary - Females",
"Middle - Persons",
"Middle - Males",
"Middle - Females",
"Matric/Secondary - Persons",
"Matric/Secondary - Males",
"Matric/Secondary - Females",
"Higher second/Interm/Pre-University/Senior second - Persons",
"Higher second/Interm/Pre-University/Senior second - Males",
"Higher second/Interm/Pre-University/Senior second - Females",
"Non-technical diploma - Persons",
"Non-technical diploma - Males",
"Non-technical diploma - Females",
"Technical diploma - Persons",
"Technical diploma - Males",
"Technical diploma - Females",
"Graduate_Persons",
"Graduate_Males",
"Graduate_Females",
"Unclassified_Persons",
"Unclassified_Males",
"Unclassified_Females"
]

# Mapeamos cada sufijo con su denominador correspondiente
denominator_map = {
    "Persons": "Total - Persons",
    "Males":   "Total - Males",
    "Females": "Total - Females"
}

# Creamos columnas de tasa con el denominador correcto según el género
for col in columns:
    # Detectamos el separador (" - " o "_") para extraer el sufijo de género
    if " - " in col:
        base, gender = col.rsplit(" - ", 1)
    else:
        base, gender = col.rsplit("_", 1)

    denominator = denominator_map[gender]
    new_col = base.lower().replace(" ", "_").replace("/", "_") + f"_{gender.lower()}_rate"
    df_Educ_state[new_col] = df_Educ_state[col] / df_Educ_state[denominator]

# Resultado final
df_Educ_state.head()
```

Merge definitivo: df_Educ_state con df_Rank_state

```python
# Merge definitivo: contiene df_Educ_state y df_Rank_state
df_merged = df_Educ_state.merge(
    df_Rank_state,
    left_on='Area Name',
    right_on='State',
    how='left'
)

df_merged['tiene_ranking'] = df_merged['Score_mean'].notna()

print(f"Estados únicos: {df_merged['Area Name'].nunique()}")  # 35
print(f"Estados con ranking: {df_merged[df_merged['tiene_ranking']]['Area Name'].nunique()}")  # 22 (Telangana queda fuera del analisis)
```

Con esto obtengo un dataframe basado en estadísticas de los diferentes niveles educativos diferenciados por estados y los promedios de los principales indicadores educativos del NIRF de los estados de la India.
